```python
import re
import datetime

# Read the RTF file and extract its content
rtf_content = ""
with open('RESULTADOS_MES/Dias_pronosticados_lista.rtf', 'r', encoding='utf-8') as rtf_file:
    rtf_content = rtf_file.read()

# Extract date strings using regular expressions
date_pattern = r'\d{4}-\d{2}-\d{2}'
date_strings = re.findall(date_pattern, rtf_content)

# Clean and parse the date strings
parsed_dates = []

for date_str in date_strings:
    try:
        # Parse the date string into a datetime object
        date_obj = datetime.datetime.strptime(date_str, '%Y-%m-%d')
        # Append the formatted date to the list
        formatted_date = date_obj.strftime("datetime.date(%Y, %m, %d)")
        parsed_dates.append(formatted_date)
    except ValueError:
        # Handle invalid date strings
        print(f"Invalid date string: {date_str}")

# Join the cleaned and formatted date strings
formatted_dates_str = ',\n'.join(parsed_dates)

print(formatted_dates_str)
```
